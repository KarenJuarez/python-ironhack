```python
import requests
import pandas as pd
url= 'https://www.merca20.com/para-competir-con-fedex-y-ivoy-uber-presenta-una-nueva-marca-de-envios-en-cuarentena/'
```

```python
#Esto nos va a traer el HTML sin estructura
html = requests.get(url).content   
```

```python
html[0:600]
```

```python
#Para limpiar el contenido que nos arrojó ya analizar mejor el texto, utilizaremos otra libreria
#BeautifulSoup, nos va a estructurar el código HTML

from bs4 import BeautifulSoup

soup=BeautifulSoup(html,"lxml")
soup
```

```python
tags=['h1','h2','h3','h4','h5','h6','h7','p'] #Nos va a traer los encabezados y parrafos que encuentre en el código 
```

```python
text=[element.text for element in soup.find_all(tags)] #Nos trae todos los elementos de HTML que corresponden a las tags como elementos de una lista.
text
```

```python
#Para ver el texto en forma de parrafo y no como una lista, usaremos el método join
print('\n'.join(text))
```

```python
url='https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
html=requests.get(url).content
soup=BeautifulSoup(html,"lxml")
```

```python
#Extraemos la tabla 
#En este caso, en la fuente de la página determinamos la clase de la tabla (sortable wikitable)
#Y el índice 0 para obtener solo la tabla que queremos
table=soup.find_all('table',{'class':'sortable wikitable'})[0]
table
```

```python
#Para cargar los datos en un Data frame necesitamos extraer las filas y sus valores de celdas en una lista anidada.
rows=table.find_all('tr') #definimos los renglones en la tabla
rows=[row.text.strip().split("\n") for row in rows]
rows
```

```python
#En esta lista anidada podemos especificarlos nombres de las columnas

colnames=rows[0]
data=rows[1:]

df=pd.DataFrame(data,columns=colnames)
```
