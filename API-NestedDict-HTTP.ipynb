```python
#En ocaiones tenemos data en JSON anidados
response = requests.get('https://api.github.com/events')
data = pd.DataFrame(response.json())
data
```

```python
#Necesitamos extraer la información anidada en diccionarios en las columnas
#Convierta los diccionarios anidados en un marco de datos con una columna para cada clave
#Asigne nombres de columna a cada columna en este nuevo marco de datos
#Agregue estas nuevas columnas al marco de datos original
#Suelte la columna con los diccionarios anidados

def flatten(data, col_list): #Vamos a crear una función para "desanidar la información"
    for column in col_list:  #Hacemos un bucle para la lista con los nombres de las columnas "col_list" 
        flattened = pd.DataFrame(dict(data[column])).transpose()   #Trasponemos las columnas del diccionario y lo convertimos en dataFrame
        columns = [str(col) for col in flattened.columns]          #Tranformamos cada columna en string y aplicamos la función creada a las columnas 
        flattened.columns = [column + '_' + colname for colname in columns] #Concatenamos la columna y elnombre de la columna 
        data = pd.concat([data, flattened], axis=1) 
        data = data.drop(column, axis=1)
    return data
```

```python
#Ahora que tenemos nuestra función, apliquémosla a las columnas que tienen diccionarios anidados y recuperemos un marco de datos revisado.

nested_columns = ['actor', 'org', 'payload', 'repo']
flat = flatten(data, nested_columns)
flat.head(10)
```
