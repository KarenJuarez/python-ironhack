# Submodulo Interpolate

```python
from scipy.interpolate import interp1d
```

```python
# Vamos a crear un conjunto de datos aleatorios entre el 1 y el 10
# Y y serán los cosenos de estos valores
x=np.arange(0,11,1)
y=np.cos(x)
```

```python
# Vamos a generar la función de interpolación y aplicaremos la función a nuestros nuevos datos y graficamos
f=interp1d(x,y)
xnew=np.arange(0,10.25,0.25)
plt.plot(xnew, f(xnew))
```

```python
xnew
```

```python
# También podemosaplicar una interpolacion spline cuadratica
```

```python
f=interp1d(x,y,'quadratic')
xnew= np.arange(0,10.25,0.25)
plt.plot(xnew,f(xnew))
```
